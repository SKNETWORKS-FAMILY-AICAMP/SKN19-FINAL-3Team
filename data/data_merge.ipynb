## 요약 학습용 데이터 생성

```python
folder_path = Path("docs_data/test_data/")
file_names = [f.stem for f in folder_path.glob("*.jsonl")]

output_file = Path("docs_data/merged_data/summary_data.jsonl")
output_file.parent.mkdir(parents=True, exist_ok=True)

total_count = 0

with output_file.open("w", encoding="UTF-8") as fout:
    for FILE_NAME in file_names:
        file_path = Path(f"docs_data/indexing_data/index_data_{FILE_NAME}.jsonl")
        file_path2 = Path(f"docs_data/test_data/{FILE_NAME}.jsonl")

        if not file_path.exists() or not file_path2.exists():
            continue

        # 1️⃣ index_data를 text 기준 dict로 로딩
        index_map = {}

        with file_path.open("r", encoding="UTF-8") as fin:
            for line in fin:
                line = line.strip()
                if not line:
                    continue
                row = json.loads(line)
                text = row.get("text")
                if text:
                    index_map[text] = row

        # 2️⃣ test_data(summary)와 병합
        with file_path2.open("r", encoding="UTF-8") as f:
            for line in f:
                line = line.strip()
                if not line:
                    continue

                row = json.loads(line)
                text = row.get("text")

                if text in index_map:
                    merged = {
                        "text": text,
                        "index": index_map[text].get("index"),
                        "summary": row.get("summary")
                    }
                    fout.write(json.dumps(merged, ensure_ascii=False) + "\n")
                    total_count += 1

print(total_count)
```
